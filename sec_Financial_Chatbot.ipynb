```python
## INTERACTIVE FUNCTIONS TO ADD TO YOUR EXISTING NOTEBOOK

def interactive_chatbot_simple():
    """
    Creates the interactive input box you want
    Works with your existing simple_chatbot() function
    """

    print("ü§ñ INTERACTIVE FINANCIAL CHATBOT")
    print("=" * 50)
    print("Ask me about Microsoft, Apple, and Tesla financial data!")
    print("Type 'help' for available queries or 'quit' to exit.")
    print("-" * 50)

    while True:
        try:
            # THIS CREATES THE INTERACTIVE INPUT BOX
            user_input = input("\nüí¨ Your question: ").strip()

            if user_input.lower() in ['quit', 'exit', 'bye']:
                print("\nüëã Thank you for using the Financial Chatbot!")
                break

            if not user_input:
                print("Please enter a question or type 'quit' to exit.")
                continue

            # Uses YOUR existing simple_chatbot() function
            response = simple_chatbot(user_input)
            print(f"\nü§ñ {response}")

        except KeyboardInterrupt:
            print("\nüëã Chatbot session ended.")
            break
        except Exception as e:
            print(f"‚ö†Ô∏è Error: {e}")

def interactive_chatbot_menu():
    """
    Alternative: Menu-style interaction (no typing required)
    Works with your existing simple_chatbot() function
    """

    print("ü§ñ INTERACTIVE FINANCIAL CHATBOT - MENU STYLE")
    print("=" * 60)

    # These match your existing queries
    queries = {
        '1': 'What is the total revenue?',
        '2': 'How has net income changed over the last year?',
        '3': 'Which company has the highest revenue?',
        '4': 'What are the growth rates?',
        '5': 'Which company is performing best?'
    }

    while True:
        print("\nüìã SELECT A QUERY:")
        print("-" * 30)
        for key, query in queries.items():
            print(f"{key}. {query}")
        print("0. Quit")

        try:
            choice = input("\nüî¢ Enter your choice (0-5): ").strip()

            if choice == '0':
                print("\nüëã Thank you for using the Financial Chatbot!")
                break
            elif choice in queries:
                selected_query = queries[choice]
                print(f"\nüí¨ You selected: {selected_query}")
                print("-" * 50)

                # Uses YOUR existing simple_chatbot() function
                response = simple_chatbot(selected_query)
                print(f"ü§ñ Response:\n{response}")
                print("\n" + "=" * 60)
            else:
                print("‚ùå Invalid choice. Please enter a number from 0-5.")

        except KeyboardInterrupt:
            print("\nüëã Chatbot session ended.")
            break

def quick_demo():
    """
    Quick demonstration using your existing chatbot
    """

    print("üöÄ QUICK CHATBOT DEMO")
    print("=" * 40)

    demo_queries = [
        "What is the total revenue?",
        "Which company has the highest revenue?",
        "What are the growth rates?"
    ]

    for i, query in enumerate(demo_queries, 1):
        print(f"\nüìã Demo {i}: {query}")
        print("-" * 40)

        # Uses YOUR existing simple_chatbot() function
        response = simple_chatbot(query)
        print(f"ü§ñ Response: {response}")

        if i < len(demo_queries):
            input("\nPress Enter to continue...")

    print("\nüéâ Demo complete!")
```
